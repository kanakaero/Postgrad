## Inverse of a Matrix

Given a matrix $A$ such that,

\begin{equation*}
A = \begin{bmatrix}
a & b \\
c & d
\end{bmatrix}
\end{equation*}  

The inverse of $A$ is given by $A^{-1}$, provided that the determinant $ad - bc \neq 0$. It is given by,
\begin{equation*}
A^{-1} = \frac{1}{ad - bc} \begin{bmatrix}
d & -b \\
-c & a
\end{bmatrix}
\end{equation*}

__Assignment 1:__ Find the inverse of the matrix $A$ given below,
\begin{equation*}
A = \begin{bmatrix}
4 & 7 \\
2 & 6
\end{bmatrix}
\end{equation*}

### With Built-in Functions

```python
import numpy as np

# Creating the Matrix
print("Inputted Matrix:")
a = float(4)
b = float(7)
c = float(2)
d = float(6)

# Converting inputs to the matrix
A = np.array([[a, b], [c, d]])

# Display the matrix
print("[",a, b,"],\n","[",c, d,"]")
print()

# Calculate its Determinant
det = (a*d) - (b*c)

# Check whether inverse exists
if det == 0:
    print("The matrix is singular and does not have an inverse.")

else:
    print("The determinant of the matrix is:", det)
    print()
    
    # Calculate the inverse of the matrix
    inverse = np.linalg.inv(A)
    print("The inverse of the matrix is:")
    print(inverse)

```
